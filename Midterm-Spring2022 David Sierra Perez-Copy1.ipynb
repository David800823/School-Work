The Speedy Internet Service Company, an ISP, charges for connectivity based on the speed of the connection and the distance the customer is from the ISP's location.<br>

The charges based on speed are as follows:<br>
   
Low-speed service: \\$50/month.<br>
High-speed service: \\$90/month.<br>

Validate the type of service that was entered.     
    
The distance charge is \\$10 per mile; customers must be within 10 miles of Speedy's location.
Validate the distance that was entered. 
    
In addition, there is a one-time installation charge of $50 and a monthly maintenance fee of \\$20/month.

Write a program which calculates the charge for the first month and the recurring monthly charge.

Sample output follows:


```python
#1 The Speedy Internet Service Company
#The distance charge is $10 per mile;
#customers must be within 10 miles of Speedy's location. 
#Validate the distance that was entered.
#In addition, there is a one-time installation charge of $50 and a monthly maintenance fee of $20/month.

distance_charge = 10
monthly_maintence = 20
installation_fee = 50 
speedy_location = 11

while speedy_location >= 10:
    speedy_location = int(input('Enter distance from our officee: '))

flag = True

while True:
    line_speed = input('What is your line speed (high or low): ')
    if line_speed == "high":
        service = 90
        break
    elif line_speed == "low":
        service = 50
        break
            
starting_charge = service+(speedy_location*distance_charge)+monthly_maintence+installation_fee
monthly_charge = service+(speedy_location*distance_charge)+monthly_maintence

print()
print(f'Charge for the first month ${starting_charge:.2f}.')
print(f'Recurring monthly charge is ${monthly_charge:.2f}.')
```
