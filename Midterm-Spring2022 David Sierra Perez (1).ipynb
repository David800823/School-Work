<li> Question #4: [15 points] Summing Pairs of Numbers<br>

Write a program which asks the user to enter two numbers and sums them, and then loops, asking the user for the next pair of numbers.

If both numbers are 0, that indicates that the data entry is complete, in which case the total of all the numbers should be printed.

Sample output follows. 

```python
#4 Summing Pairs of Numbers

num1 = 1
num2 = 1
sum_of_all = 0

while num1 != 0 and num2 != 0:
    num1 = int(input('Enter a number: '))
    num2 = int(input('Enter another number: '))
    total_num = num1 + num2
    sum_of_all += total_num 
    if num1 ==0 and num2 ==0:
        continue
    print(f'The sum of {num1:.1f} and {num2:.1f} is {total_num:.1f}')

print(f'The sum of all the numbers is {sum_of_all:.1f}')
```
